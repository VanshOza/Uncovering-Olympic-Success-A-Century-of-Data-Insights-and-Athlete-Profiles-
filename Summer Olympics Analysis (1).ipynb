```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt

df_clean = df.dropna(subset=['Sex', 'NOC', 'City', 'Sport', 'Event', 'HOST_COUNTRY', 'Age'])

df_clean['Medal_Binary'] = df_clean['Medal'].notna().astype(int)

X_clean = df_clean.drop(['Medal', 'Team', 'Name', 'Games', 'Year', 'Season', 'Medal_Binary'], axis=1)
Y_clean = df_clean['Medal_Binary']

X_encoded = pd.get_dummies(X_clean, columns=['Sex', 'NOC', 'City', 'Sport', 'Event', 'HOST_COUNTRY'])

X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y_clean, test_size=0.2, random_state=42, stratify=Y_clean)

model = DecisionTreeClassifier(
    max_depth=3,          
    min_samples_split=50, 
    class_weight='balanced',
    random_state=42
)
model.fit(X_train, Y_train)

predictions = model.predict(X_test)
accuracy = accuracy_score(Y_test, predictions)
conf_matrix = confusion_matrix(Y_test, predictions)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)

# Plot decision tree
plt.figure(figsize=(15, 7))
plot_tree(model, feature_names=X_encoded.columns, class_names=['No Medal', 'Medal'], filled=True, rounded=True)
plt.title("Simplified Decision Tree")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer

df_clean = df.dropna(subset=['Sex', 'NOC', 'City', 'Sport', 'Event', 'HOST_COUNTRY'])

Y_clean = df_clean['Medal'].replace({np.nan: 0, 'Gold': 1, 'Silver': 1, 'Bronze': 1}).astype(int)

X_clean = df_clean.drop(['Medal', 'Team', 'Name', 'Games', 'Year', 'Season'], axis=1)

X_encoded = pd.get_dummies(X_clean, columns=['Sex', 'NOC', 'City', 'Sport', 'Event', 'HOST_COUNTRY'], drop_first=True)

# Imputation to handle missing values
imputer = SimpleImputer(strategy='most_frequent')  # Using the most frequent value for imputation
X_imputed = imputer.fit_transform(X_encoded)

# Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X_imputed, Y_clean, test_size=0.2, random_state=42)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Model Definitions
models = {
    'Logistic Regression': LogisticRegression(class_weight='balanced', random_state=42),
    'Decision Tree': DecisionTreeClassifier(class_weight='balanced', random_state=42),
    'Random Forest': RandomForestClassifier(class_weight='balanced', random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier()
}

# Model Evaluation
model_scores = {}

for model_name, model in models.items():
    print(f"Training and evaluating {model_name}...")
    
    if model_name in ['Logistic Regression', 'Decision Tree', 'Random Forest']:
        model.fit(X_train_scaled, Y_train)  
    else:
        model.fit(X_train, Y_train)  
    
    prediction = model.predict(X_test_scaled if model_name != 'K-Nearest Neighbors' else X_test)
    
    score = accuracy_score(Y_test, prediction)
    model_scores[model_name] = score
    print(f"{model_name} Accuracy: {score:.4f}")

print("\nAll Model Accuracies:")
for model_name, score in model_scores.items():
    print(f"{model_name}: {score:.4f}")
```
